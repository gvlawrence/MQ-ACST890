```python
group_mean = dataset[["Class No", "Total_marks"]]. groupby('Class No').mean()
group_var = dataset[["Class No", "Total_marks"]]. groupby('Class No').var()
group_std = dataset[["Class No", "Total_marks"]]. groupby('Class No').std()
stat_table = pd.merge(pd.merge(group_mean, group_var, on= 'Class No'), group_std, on = 'Class No')
stat_table.head()
```

```python
raw_data = {'ID number': [dataset['ID number']], 'Total_marks': [dataset['Total_marks']]}
df = pd.DataFrame(raw_data, columns = ['ID number', 'Total_marks'])

# Create a list to store the data
Grade = []

# For each row in the column,
for row in df['Total_marks']:
    # if more than a value,
    if row > 95:
        # Append a letter grade
        Grade.append('A')
    # else, if more than a value,
    elif row > 90:
        # Append a letter grade
        Grade.append('A-')
    # else, if more than a value,
    elif row > 85:
        # Append a letter grade
        Grade.append('B')
    # else, if more than a value,
    elif row > 80:
        # Append a letter grade
        Grade.append('B-')
    # else, if more than a value,
    elif row > 75:
        # Append a letter grade
        Grade.append('C')
    # else, if more than a value,
    elif row > 70:
        # Append a letter grade
        Grade.append('C-')
    # else, if more than a value,
    elif row > 65:
        # Append a letter grade
        Grade.append('D')
    # else, if more than a value,
    elif row > 60:
        # Append a letter grade
        Grade.append('D-')
    # otherwise,
    else:
        # Append a failing grade
        Grade.append('Failed')
        
# Create a column from the list
df['Grade'] = Grade
```
